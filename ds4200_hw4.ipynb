```python
import pandas as pd
```

```python
data = pd.read_csv('us_counties_covid19_daily1.csv')
```

```python
data
```

```python
from datetime import datetime
dates = data['date'].to_numpy()
date_converted = []
for i in dates:
    date_converted.append(datetime.strptime(i, '%Y-%m-%d'))
data['date'] = date_converted
data.set_index('date', inplace = True)
```

```python
data
```

```python
newcase = data[['cases']].copy()
newcase = newcase.groupby('date').sum()
newcase.to_csv('newcase.csv')
newcase
```

```python
increase = newcase.resample('Q').sum()
increase.iloc[3] = increase.iloc[3] - increase.iloc[2]
increase.iloc[2] = increase.iloc[2] - increase.iloc[1]
increase.iloc[1] = increase.iloc[1] - increase.iloc[0]
increase.to_csv('increase.csv')
increase
```

```python
newdeath = data[['deaths']].copy()
newdeath = newdeath.groupby('date').sum()
newdeath.to_csv('newdeaths.csv')
newdeath
```

```python
increaseD = newdeath.resample('Q').sum()
increaseD.iloc[3] = increaseD.iloc[3] - increaseD.iloc[2]
increaseD.iloc[2] = increaseD.iloc[2] - increaseD.iloc[1]
increaseD.iloc[1] = increaseD.iloc[1] - increaseD.iloc[0]
increaseD.to_csv('increaseD.csv')
increaseD
```

```python
tmp = newcase.copy()

length = len(newcase)
while length - 1 > 0:
    tmp.iloc[length - 1] = tmp.iloc[length - 1] - tmp.iloc[length - 2]
    length -= 1
tmp.to_csv('changeday.csv')
tmp
```
